```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
image = cv2.imread("C:/Users/USER/Desktop/Machine_learning/Image_to_pencil_sketch/Zayn.png")
```

```python
uc_image =cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
invert_image = 255 - uc_image
```

```python
blurred_image = cv2.GaussianBlur(invert_image,(21,21),0)
```

```python
invert_blur_image = 255 - blurred_image
```

```python
pencil_sketch_image = cv2.divide(uc_image,invert_blur_image, scale=256.0)
```

```python
cv2.imshow('Original Image',image)
plt.imshow(image)
```

```python
cv2.imshow('B&W Image',uc_image)
plt.imshow(uc_image)
```

```python
cv2.imshow('Inverted Image',invert_image)
plt.imshow(invert_image)
```

```python
cv2.imshow('Blurred Image',blurred_image)
plt.imshow(blurred_image)
```

```python
cv2.imshow('Inverted Blurred Image',invert_blur_image)
plt.imshow(invert_blur_image)
```

```python
cv2.imshow('Pencil Sketch Image',pencil_sketch_image)
plt.imshow(pencil_sketch_image)
```
